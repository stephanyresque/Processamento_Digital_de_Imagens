```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
img = cv.imread('Exemplo_Negativar.jpg', cv.IMREAD_GRAYSCALE)
```

```python
negativo = ~img
```

```python
print(f'Dimensões: {img.shape[0]}x{img.shape[1]}')

print("Imagem original")
cv2_imshow(img)

print("Imagem negativa")
cv2_imshow(negativo)
```

```python
img_a = cv.imread("Exemplo_Subtrair_A.jpg")
img_b = cv.imread("Exemplo_Subtrair_B.jpg")
```

```python
print(f"Dimensões imagem A: {img_a.shape[0]}x{img_a.shape[1]}X{img_a.shape[2]}")
print(f"Dimensões imagem B: {img_b.shape[0]}x{img_b.shape[1]}X{img_b.shape[2]}")

subtracao = cv.subtract(img_b, img_a)

print('Imagem A:')
cv2_imshow(img_a)

print('\n\nImagem B:')
cv2_imshow(img_b)

print('\n\nSubtração:')
cv2_imshow(subtracao)
```

```python
img = cv.imread("Exemplo_Threshold.bmp", cv.IMREAD_GRAYSCALE)
```

```python
print(f'Dimensões: {img.shape[1]}x{img.shape[0]}')

L = 150

_,threshold = cv.threshold(img, L, 255, cv.THRESH_BINARY)

print("Imagem orginal")
cv2_imshow(img)

print('\n\nImagem limiarizada:')
cv2_imshow(threshold)
```

```python
img = cv.imread("Exemplo_Threshold.jpg", cv.IMREAD_GRAYSCALE)
```

```python
print(f'Dimensões: {img.shape[1]}x{img.shape[0]}')

L = 100

_,threshold = cv.threshold(img, L, 255, cv.THRESH_BINARY)

print("Imagem orginal")
cv2_imshow(img)

print('\n\nImagem limiarizada:')
cv2_imshow(threshold)
```

```python
img = cv.imread("Exemplo_Threshold.jpg", cv.IMREAD_GRAYSCALE)
```

```python
print(f"Dimensões: {img.shape[1]}x{img.shape[0]}")

mean_thres = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 15, 10)
gaussian_thres = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 15, 10)

print('Imagem original:')
cv2_imshow(img)
print('\n\nLimiarização adaptativa pela media:')
cv2_imshow(mean_thres)
print('\n\nLimiarização adaptativa pela gaussiana:')
cv2_imshow(gaussian_thres)
```
