```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('Exemplo_Segmentacao_0.jpg')
```

```python
print('Original')
cv2_imshow(img)

imagem_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

cv2.line(img, (0, 142), (img.shape[1], 142), (0, 255, 0), 1)
print('Original-Destaque')
cv2_imshow(img)

print()

eixo_x = range(imagem_gray.shape[1])
valores = []
for i in range(imagem_gray.shape[1]):
    valores.append(int(imagem_gray[142,i]))

plt.plot(eixo_x, valores)
plt.show()
```
